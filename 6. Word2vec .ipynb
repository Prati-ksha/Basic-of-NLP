data pre processing

```python
text= re.sub(r'\[[0-9]*\]', ' ', para)
text= re.sub(r'\s+', ' ', text)
text= text.lower()
text= re.sub(r'\d', ' ', text)
text= re.sub(r'\s+', ' ' , text)
```

```python
print(text)
```

preparing the dataset

```python
sen= nltk.sent_tokenize(text)
sen= [nltk.word_tokenize(sentence) for sentence in sen]
for i in range (len(sen)):
  sen[i]= [word for word in sen[i] if word not in stopwords.words('english')]
```

creating the model

```python
model= Word2Vec(sen, min_count=1)
x = list(model.wv.index_to_key)
vector= model.wv['bts']
sim= model.wv.most_similar(['bts'])
```

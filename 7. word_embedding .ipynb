```python
from keras.preprocessing.text import one_hot
```

```python
sent= ['i am a girl',
       'i love anime',
       'the milk is hot',
       'the coffee is bitter',
       'apple is red in colour',
       'ai is evovling everyday',
       'a black hole is very beautiful',
       'organic chemistry is all about carbon',
       'cats are cute',
       'luffy is one of the best main lead']
```

```python
sent
```

```python
vocab= 7000
```

```python
on_hot= [one_hot(word ,vocab) for word in sent]
```

```python
print(on_hot)
```

```python
from keras.models import Sequential
from keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
```

```python
sent_length= 8
emb= pad_sequences(on_hot, padding='post', maxlen= sent_length)
```

```python
emb
```

```python
dim= 10
model= Sequential()
model.add(Embedding(vocab, 10, input_length= sent_length))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
print(model.predict(emb))
```
